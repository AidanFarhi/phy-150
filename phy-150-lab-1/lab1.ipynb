```python
import statistics
import matplotlib.pyplot as plt
```

```python
def graph_data(x, y, title, x_label, y_label):
    plt.plot(x, y, marker='o', linestyle='-')
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True)
    plt.show()
```

Given Data

```python
given_x = [0, 5, 10, 15, 20, 30, 35, 40, 45, 50]
given_y = [0, 20, 40, 50, 55, 60, 70, 70, 70, 55]
```

Distance vs Time of a Moving Object - Graph

```python
graph_data(given_x, given_y, 'Distance vs Time of a Moving Object', 'Time (s)', 'Position (m)')
```

Constant Velocity Vehicle Data

```python
red_buggy_x = [0, 0.47, 1.27, 1.84, 2.43, 3.12, 3.79, 4.56, 5.00]
red_buggy_y = [0, 0.25, 0.50, 0.75, 1.0, 1.25, 1.50, 1.75, 2.00]
```

```python
graph_data(red_buggy_x, red_buggy_y, 'Buggy Displacement vs Time', 'Time (s)', 'Displacement (m)')
```

Generate Velocity Table

```python
velocity_vals = []
for i in range(1, len(red_buggy_x)):
    v = (red_buggy_y[i] - red_buggy_y[i-1]) / (red_buggy_x[i] - red_buggy_x[i-1])
    velocity_vals.append((i, round(v, 2)))
```

```python
vel_x = [x[0] for x in velocity_vals]
vel_y = [x[1] for x in velocity_vals]
```

```python
graph_data(vel_x, vel_y, 'Buggy Velocity vs Time', 'Time (s)', 'Velocity (m/s)')
```

Sphere Data

```python
sphere_times = [
    (0.74, 0.47, 0.45), (1.05, 0.74, 0.70), (1.18, 0.90, 0.90),
    (1.10, 1.30, 1.31), (1.35, 1.53, 1.42), (1.71, 1.62, 1.65),
    (2.01, 1.94, 1.96), (2.32, 2.41, 2.52)
]
sphere_y = [x / 10 for x in range(1, 9)]
sphere_x = [round(statistics.mean(x), 2) for x in sphere_times]
sphere_squared_x = [round(x ** 2, 2) for x in sphere_x]
```

```python
graph_data(sphere_x, sphere_y, 'Shpere Distance vs Time', 'Time (s)', 'Distance (m)')
```

```python
graph_data(sphere_squared_x, sphere_y, 'Shpere Distance vs Time Squared', 'Time (s) Squared', 'Distance (m)')
```
